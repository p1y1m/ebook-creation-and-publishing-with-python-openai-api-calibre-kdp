```python
# ============================================
# Build EPUB (12 chapters) + auto-pair images
# Layout: Title (EN/ES) -> Image (fit) -> Content
# ============================================

!pip -q install ebooklib markdown

from google.colab import drive
drive.mount("/content/drive")

import os, re
from datetime import datetime
from ebooklib import epub
import markdown as md

# ----------------------------
# Configure paths
# ----------------------------
BOOK_DIR = "/content/drive/MyDrive/Ebooks/CBT Anxiety for Kids"
MD_PATH  = f"{BOOK_DIR}/cbt_anxiety_skills_for_kids_Fixedv4.md"
IMG_DIR  = f"{BOOK_DIR}/images"

TITLE  = "CBT Anxiety Skills for Kids / Habilidades CBT para la Ansiedad en Niños"
AUTHOR = "Pedro Yanez Melendez"

ts = datetime.now().strftime("%Y%m%d_%H%M%S")
OUT_EPUB = f"{BOOK_DIR}/CBT_Anxiety_Skills_for_Kids_{ts}.epub"

assert os.path.exists(MD_PATH),  f"Markdown file not found: {MD_PATH}"
assert os.path.exists(IMG_DIR),  f"Images folder not found: {IMG_DIR}"

# ----------------------------
# Read markdown
# ----------------------------
with open(MD_PATH, "r", encoding="utf-8") as f:
    raw = f.read().replace("\r\n", "\n")

# ----------------------------
# Parse chapters robustly (accept '### Chapter' or 'Chapter')
# ----------------------------
chapter_re = re.compile(r"^(?:###\s*)?Chapter\s+(\d+):\s*(.+)\s*$", re.MULTILINE)
matches = list(chapter_re.finditer(raw))

if not matches:
    raise ValueError("No chapter headings found. Expect lines like: '### Chapter 1: ...' or 'Chapter 1: ...'")

# Slice intro (optional)
intro_text = raw[:matches[0].start()].strip()

chapters = []
for idx, m in enumerate(matches):
    ch_num = int(m.group(1))
    ch_title = m.group(0).strip()  # full bilingual title line as written in file
    start = m.end()
    end = matches[idx + 1].start() if idx + 1 < len(matches) else len(raw)
    ch_body = raw[start:end].strip()
    chapters.append((ch_num, ch_title, ch_body))

# Enforce correct order and count
chapters.sort(key=lambda x: x[0])

if len(chapters) != 12:
    raise ValueError(f"Expected 12 chapters, found {len(chapters)}. Check chapter headings in the .md file.")

# ----------------------------
# Create EPUB book
# ----------------------------
book = epub.EpubBook()
book.set_identifier("cbt-anxiety-kids")
book.set_title(TITLE)
book.set_language("en")
book.add_author(AUTHOR)

# ----------------------------
# Add CSS (fit inside images; title above image)
# ----------------------------
STYLE = """
body {
  font-family: serif;
  padding: 4vh 6vw;
  line-height: 1.45;
}

h1, h2 {
  text-align: center;
  margin: 0.8em 0 0.4em 0;
}

img.chapter-illus {
  max-width: 100%;
  max-height: 70vh;
  object-fit: contain;
  display: block;
  margin: 0 auto 2vh auto;
}

hr { display: none; } /* hide markdown separators if any become <hr> */
"""

css_item = epub.EpubItem(uid="style", file_name="style.css", media_type="text/css", content=STYLE)
book.add_item(css_item)

# ----------------------------
# Convert markdown to HTML cleanly (no raw ###, -, ---)
# ----------------------------
def md_to_html(md_text: str) -> str:
    # Convert using python-markdown (handle lists/bold nicely)
    html = md.markdown(
        md_text,
        extensions=["extra", "sane_lists"]
    )
    # Remove any <hr> produced from --- separators (already hidden via CSS, but also remove)
    html = re.sub(r"<hr\s*/?>", "", html, flags=re.IGNORECASE)
    return html

# ----------------------------
# Find matching image by prefix 'chapter_01', 'chapter_02', ...
# Prefer latest if multiple exist
# ----------------------------
def find_chapter_image(ch_num: int):
    prefix = f"chapter_{ch_num:02d}"
    candidates = sorted([f for f in os.listdir(IMG_DIR) if f.startswith(prefix)])
    if not candidates:
        return None, None, None

    filename = candidates[-1]  # pick last = newest lexicographically (timestamped)
    path = os.path.join(IMG_DIR, filename)

    ext = os.path.splitext(filename)[1].lower()
    if ext == ".png":
        media_type = "image/png"
    elif ext in [".jpg", ".jpeg"]:
        media_type = "image/jpeg"
    else:
        # default (still usually works)
        media_type = "image/png"
    return filename, path, media_type

# ----------------------------
# Build intro page (optional)
# ----------------------------
epub_items = []
spine_items = ["nav"]

if intro_text:
    intro_html = f"""
    <html><head><link rel="stylesheet" href="style.css" type="text/css"/></head>
    <body>
      {md_to_html(intro_text)}
    </body></html>
    """
    intro = epub.EpubHtml(title="Intro", file_name="intro.xhtml", lang="en")
    intro.content = intro_html
    book.add_item(intro)
    epub_items.append(intro)
    spine_items.append(intro)

# ----------------------------
# Build chapters with strict mapping: chapter N -> image prefix chapter_NN
# Layout: Title -> Image -> Content
# ----------------------------
for ch_num, ch_title_line, ch_body in chapters:
    # Prepare title (keep bilingual title exactly as in file)
    title_html = f"<h2>{ch_title_line}</h2>"

    # Add image if found
    img_filename, img_path, img_media_type = find_chapter_image(ch_num)
    img_html = ""
    if img_filename:
        with open(img_path, "rb") as f:
            img_bytes = f.read()

        img_item = epub.EpubItem(
            uid=f"img_{ch_num:02d}",
            file_name=img_filename,
            media_type=img_media_type,
            content=img_bytes
        )
        book.add_item(img_item)

        img_html = f'<img class="chapter-illus" src="{img_filename}" alt="Chapter illustration"/>'
    else:
        # Keep empty if missing, but do not break chapter alignment
        img_html = ""

    # Convert chapter body markdown to HTML
    body_html = md_to_html(ch_body)

    # Compose final XHTML content
    chapter_xhtml = f"""
    <html>
      <head><link rel="stylesheet" href="style.css" type="text/css"/></head>
      <body>
        {title_html}
        {img_html}
        {body_html}
      </body>
    </html>
    """

    chap = epub.EpubHtml(
        title=f"Chapter {ch_num:02d}",
        file_name=f"chapter_{ch_num:02d}.xhtml",
        lang="en"
    )
    chap.content = chapter_xhtml

    book.add_item(chap)
    epub_items.append(chap)
    spine_items.append(chap)

# ----------------------------
# TOC + spine
# ----------------------------
# Use actual chapter titles in TOC
book.toc = tuple(epub_items)
book.spine = spine_items

book.add_item(epub.EpubNcx())
book.add_item(epub.EpubNav())

# ----------------------------
# Write EPUB
# ----------------------------
epub.write_epub(OUT_EPUB, book)

print("✅ EPUB generated successfully")
print("Saved to:", OUT_EPUB)
print("Images folder used:", IMG_DIR)
print("Markdown used:", MD_PATH)
```

```python
# ============================================
# FINAL CLEAN EPUB GENERATOR
# ============================================

!pip -q install ebooklib markdown

from google.colab import drive
drive.mount("/content/drive")

import os, re
from datetime import datetime
from ebooklib import epub
import markdown as md

# ----------------------------
# CONFIG
# ----------------------------
BOOK_DIR = "/content/drive/MyDrive/Ebooks/CBT Anxiety for Kids"
MD_PATH  = f"{BOOK_DIR}/cbt_anxiety_skills_for_kids_Fixedv4.md"
IMG_DIR  = f"{BOOK_DIR}/images"

TITLE  = "CBT Anxiety Skills for Kids / Habilidades CBT para la Ansiedad en Niños"
AUTHOR = "Pedro Yanez Melendez"

ts = datetime.now().strftime("%Y%m%d_%H%M%S")
OUT_EPUB = f"{BOOK_DIR}/CBT_Anxiety_Skills_for_Kids_{ts}.epub"

assert os.path.exists(MD_PATH),  f"Markdown file not found: {MD_PATH}"
assert os.path.exists(IMG_DIR),  f"Images folder not found: {IMG_DIR}"

# ----------------------------
# READ MARKDOWN
# ----------------------------
with open(MD_PATH, "r", encoding="utf-8") as f:
    raw = f.read().replace("\r\n", "\n")

# ----------------------------
# CHAPTER PARSER (robust)
# ----------------------------
chapter_re = re.compile(r"^(?:###\s*)?Chapter\s+(\d+):\s*(.+)$", re.MULTILINE)
matches = list(chapter_re.finditer(raw))

if len(matches) != 12:
    raise ValueError(f"Expected 12 chapters, found {len(matches)}")

chapters = []

for i, m in enumerate(matches):
    ch_num = int(m.group(1))
    title_clean = f"Chapter {ch_num}: {m.group(2)}"
    title_clean = title_clean.replace("###", "").strip()

    start = m.end()
    end = matches[i+1].start() if i+1 < len(matches) else len(raw)
    body = raw[start:end].strip()

    chapters.append((ch_num, title_clean, body))

chapters.sort(key=lambda x: x[0])

# ----------------------------
# EPUB INIT
# ----------------------------
book = epub.EpubBook()
book.set_identifier("cbt-anxiety-kids")
book.set_title(TITLE)
book.set_language("en")
book.add_author(AUTHOR)

# ----------------------------
# CLEAN CSS
# ----------------------------
STYLE = """
body {
  font-family: serif;
  padding: 5vh 6vw;
  line-height: 1.6;
}

h2 {
  text-align: center;
  margin-bottom: 2vh;
}

img.chapter-illus {
  max-width: 100%;
  max-height: 70vh;
  object-fit: contain;
  display: block;
  margin: 0 auto 3vh auto;
}

.skillbox {
  margin-top: 3vh;
  padding: 2vh;
  background-color: #f7f7f7;
  border-radius: 8px;
}
"""

css_item = epub.EpubItem(uid="style", file_name="style.css",
                         media_type="text/css", content=STYLE)
book.add_item(css_item)

# ----------------------------
# CLEAN MARKDOWN FUNCTION
# ----------------------------
def clean_markdown(md_text):

    # Remove leading ### if any
    md_text = md_text.replace("###", "")

    # Remove markdown horizontal rules
    md_text = re.sub(r"^\s*---\s*$", "", md_text, flags=re.MULTILINE)

    # Convert bullet lists into plain lines (remove - )
    md_text = re.sub(r"^\s*-\s+", "", md_text, flags=re.MULTILINE)

    # Preserve line breaks inside stories
    md_text = md_text.replace("\n", "  \n")

    html = md.markdown(md_text, extensions=["extra"])

    return html

# ----------------------------
# IMAGE FINDER
# ----------------------------
def find_image(ch_num):
    prefix = f"chapter_{ch_num:02d}"
    files = sorted([f for f in os.listdir(IMG_DIR) if f.startswith(prefix)])
    if not files:
        return None, None, None

    filename = files[-1]
    path = os.path.join(IMG_DIR, filename)

    ext = filename.split(".")[-1].lower()
    media = "image/png" if ext == "png" else "image/jpeg"

    return filename, path, media

# ----------------------------
# BUILD CHAPTERS
# ----------------------------
epub_items = []
spine_items = ["nav"]

for ch_num, ch_title, ch_body in chapters:

    img_filename, img_path, img_media = find_image(ch_num)

    img_html = ""
    if img_filename:
        with open(img_path, "rb") as f:
            img_bytes = f.read()

        img_item = epub.EpubItem(uid=f"img_{ch_num:02d}",
                                 file_name=img_filename,
                                 media_type=img_media,
                                 content=img_bytes)
        book.add_item(img_item)

        img_html = f'<img class="chapter-illus" src="{img_filename}" alt="Illustration"/>'

    body_html = clean_markdown(ch_body)

    chapter_html = f"""
    <html>
      <head>
        <link rel="stylesheet" href="style.css" type="text/css"/>
      </head>
      <body>
        <h2>{ch_title}</h2>
        {img_html}
        {body_html}
      </body>
    </html>
    """

    chap = epub.EpubHtml(title=f"Chapter {ch_num}",
                         file_name=f"chapter_{ch_num:02d}.xhtml",
                         lang="en")
    chap.content = chapter_html

    book.add_item(chap)
    epub_items.append(chap)
    spine_items.append(chap)

# ----------------------------
# TOC + SPINE
# ----------------------------
book.toc = tuple(epub_items)
book.spine = spine_items

book.add_item(epub.EpubNcx())
book.add_item(epub.EpubNav())

# ----------------------------
# EXPORT
# ----------------------------
epub.write_epub(OUT_EPUB, book)

print("✅ EPUB generated successfully")
print("Saved to:", OUT_EPUB)
```
